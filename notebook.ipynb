## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

### Assessing Data

#### Identifying Bike Sharing Dataset  problem

```python
# Cek informasi dataset
day_df.info()
hour_df.info()

# Cek missing values
print("Missing values day_df:\n", day_df.isna().sum())
print("\nMissing values hour_df:\n", hour_df.isna().sum())

# Cek data duplikat
print("\nJumlah duplikat day_df:", day_df.duplicated().sum())
print("Jumlah duplikat hour_df:", hour_df.duplicated().sum())

# Cek statistik deskriptif
day_df.describe()
hour_df.describe()
```

**Steps to Take:**

- Mengubah variabel kategorikal seperti season, weathersit, dan weekday menjadi label yang lebih deskriptif
- Memastikan kembali tidak ada missing value dan data duplikat sebelum masuk ke tahap cleaning
- Menyiapkan struktur data agar lebih mudah digunakan pada tahap Exploratory Data Analysis (EDA)

**Insight:** (Opsional)

### Cleaning Data

#### Fixing Bike Sharing Dataset  problem

```python
# Copy dataset agar aman
day_clean = day_df.copy()
hour_clean = hour_df.copy()

# 1. Convert kolom tanggal ke datetime
day_clean['dteday'] = pd.to_datetime(day_clean['dteday'])
hour_clean['dteday'] = pd.to_datetime(hour_clean['dteday'])

# 2. Mapping kategori agar lebih informatif
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
weather_map = {1: 'Clear', 2: 'Mist', 3: 'Light Snow/Rain', 4: 'Heavy Rain/Snow'}
weekday_map = {0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday',
               4: 'Thursday', 5: 'Friday', 6: 'Saturday'}

day_clean['season'] = day_clean['season'].map(season_map)
day_clean['weathersit'] = day_clean['weathersit'].map(weather_map)
day_clean['weekday'] = day_clean['weekday'].map(weekday_map)

hour_clean['season'] = hour_clean['season'].map(season_map)
hour_clean['weathersit'] = hour_clean['weathersit'].map(weather_map)
hour_clean['weekday'] = hour_clean['weekday'].map(weekday_map)

# 3. Cek ulang hasil cleaning
day_clean.info()
hour_clean.info()
```

**Insight:** (Opsional)
- Setelah dilakukan proses cleaning, tipe data pada kolom dteday telah berhasil dikonversi menjadi format datetime, sehingga memudahkan analisis berbasis waktu
- Variabel kategorikal seperti season, weathersit, dan weekday telah diubah menjadi label yang lebih deskriptif, sehingga data menjadi lebih mudah dipahami
- Tidak ditemukan missing value maupun data duplikat, sehingga tidak diperlukan penanganan tambahan pada aspek tersebut
- Dataset kini dalam kondisi lebih bersih, terstruktur, dan siap digunakan untuk tahap Exploratory Data Analysis (EDA)

## Exploratory Data Analysis (EDA)

### Explore Bike Sharing Dataset

```python
# Exploratory Data Analysis (EDA)

# Pertanyaan 1:
# Bagaimana pengaruh kondisi cuaca terhadap rata-rata jumlah penyewaan sepeda harian selama periode 2011–2012?

weather_analysis = day_clean.groupby('weathersit')['cnt'].mean().sort_values(ascending=False)
print("Rata-rata jumlah penyewaan sepeda berdasarkan kondisi cuaca:")
display(weather_analysis)

# Pertanyaan 2:
# Bagaimana perbedaan rata-rata jumlah penyewaan sepeda berdasarkan hari dalam seminggu dan musim selama periode 2011–2012?

weekday_analysis = day_clean.groupby('weekday')['cnt'].mean()
print("Rata-rata jumlah penyewaan sepeda berdasarkan hari:")
display(weekday_analysis)

season_analysis = day_clean.groupby('season')['cnt'].mean().sort_values(ascending=False)
print("Rata-rata jumlah penyewaan sepeda berdasarkan musim:")
display(season_analysis)
```

**Insight:** (Opsional)
1. Pengaruh Cuaca terhadap Penyewaan Sepeda
- Kondisi cuaca cerah (Clear) memiliki rata-rata jumlah penyewaan sepeda tertinggi
- Penyewaan cenderung menurun pada kondisi Mist dan semakin rendah saat Light Snow/Rain
- Hal ini menunjukkan bahwa cuaca sangat berpengaruh terhadap minat pengguna dalam menyewa sepeda
2. Pola Penyewaan Berdasarkan Hari
- Terdapat perbedaan jumlah penyewaan antara hari kerja dan akhir pekan
- Hari tertentu menunjukkan jumlah penyewaan yang lebih tinggi, yang mengindikasikan adanya pola penggunaan berdasarkan aktivitas harian pengguna
3. Pola Penyewaan Berdasarkan Musim
- Musim dengan cuaca lebih nyaman seperti Fall atau Summer cenderung memiliki jumlah penyewaan lebih tinggi
- Musim dengan kondisi cuaca ekstrem seperti Winter menunjukkan jumlah penyewaan yang lebih rendah
- Hal ini menunjukkan bahwa musim juga memengaruhi tingkat penggunaan layanan bike sharing

Kesimpulan Sementara:
Faktor cuaca dan waktu (hari & musim) memiliki pengaruh signifikan terhadap jumlah penyewaan sepeda
Data menunjukkan adanya pola yang konsisten dan dapat digunakan sebagai dasar pengambilan keputusan

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(8,5))
sns.barplot(x=weather_analysis.index, y=weather_analysis.values)

plt.title('Rata-rata Penyewaan Sepeda Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Rata-rata Jumlah Penyewaan')
plt.xticks(rotation=30)
plt.show()
```

### Pertanyaan 2:

```python
plt.figure(figsize=(8,5))
sns.barplot(x=weekday_analysis.index, y=weekday_analysis.values)

plt.title('Rata-rata Penyewaan Sepeda Berdasarkan Hari')
plt.xlabel('Hari')
plt.ylabel('Rata-rata Jumlah Penyewaan')
plt.xticks(rotation=30)
plt.show()
```

Berdasarkan hasil exploratory data analysis dan visualisasi, ditemukan bahwa kondisi cuaca dan faktor waktu memiliki pengaruh yang signifikan terhadap jumlah penyewaan sepeda. Penyewaan sepeda cenderung lebih tinggi pada kondisi cuaca cerah, dan menurun ketika kondisi cuaca memburuk seperti hujan atau salju ringan. Hal ini menunjukkan bahwa kenyamanan lingkungan menjadi faktor penting dalam keputusan pengguna untuk menyewa sepeda.

Selain itu, terdapat pola penyewaan berdasarkan waktu, baik dari sisi hari maupun musim. Perbedaan jumlah penyewaan antar hari menunjukkan adanya pola aktivitas pengguna, sementara musim dengan kondisi cuaca yang lebih nyaman seperti summer dan fall memiliki tingkat penyewaan yang lebih tinggi dibandingkan musim dengan kondisi ekstrem seperti winter.

Secara keseluruhan, faktor lingkungan dan waktu berperan penting dalam memengaruhi perilaku pengguna dalam menggunakan layanan bike sharing.

## Analisis Lanjutan (Opsional)

```python
# Membuat kategori tingkat penyewaan (binning)
day_clean['rental_category'] = pd.qcut(day_clean['cnt'], q=3, labels=['Low', 'Medium', 'High'])

# Melihat jumlah data tiap kategori
rental_dist = day_clean['rental_category'].value_counts()

# Rata-rata penyewaan per kategori
rental_mean = day_clean.groupby('rental_category')['cnt'].mean()

print("Distribusi kategori penyewaan:")
display(rental_dist)

print("\nRata-rata penyewaan tiap kategori:")
display(rental_mean)
```
